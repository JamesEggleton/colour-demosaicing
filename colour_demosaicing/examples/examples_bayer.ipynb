# Colour - Demosaicing - Examples: Bayer

This notebook showcase various *Bayer* CFA (Colour Filter Array) related examples.

> Note: All the showcased objects support an optional **pattern** argument specifying the arrangement of the colour filters on the pixel array: **'RGGB', 'BGGR', 'GRBG', 'GBRG'**

```python
% matplotlib inline
```

```python
from __future__ import division

import numpy as np
import os

import colour
from colour.plotting import *

from colour_demosaicing import (
    EXAMPLES_RESOURCES_DIRECTORY,
    demosaicing_CFA_Bayer_bilinear,
    demosaicing_CFA_Bayer_Malvar2004,
    demosaicing_CFA_Bayer_Menon2007,
    mosaicing_CFA_Bayer)

OECF = colour.RGB_COLOURSPACES['sRGB'].OECF
```

## Data

```python
LIGHTHOUSE_IMAGE = colour.read_image(
    os.path.join(EXAMPLES_RESOURCES_DIRECTORY, 'bayer', 'Lighthouse.exr'))

image_plot(OECF(LIGHTHOUSE_IMAGE), 'Lighthouse - R914108 - Kodak')
```

## Mosaicing

```python
CFA = mosaicing_CFA_Bayer(LIGHTHOUSE_IMAGE)

image_plot(OECF(CFA), 'Lighthouse - CFA - RGGB')

image_plot(OECF(mosaicing_CFA_Bayer(LIGHTHOUSE_IMAGE, 'BGGR')), 
           'Lighthouse - CFA - BGGR')
```

## Demosaicing - Bilinear

```python
image_plot(OECF(demosaicing_CFA_Bayer_bilinear(CFA)), 
                'Lighthouse - Demosaicing - Bilinear')
```

## Demosaicing - Malvar (2004)

```python
image_plot(OECF(demosaicing_CFA_Bayer_Malvar2004(CFA)), 
                'Lighthouse - Demosaicing - Malvar (2004)')
```

## Demosaicing - Menon (2007)

```python
image_plot(OECF(demosaicing_CFA_Bayer_Menon2007(CFA)), 
                'Lighthouse - Demosaicing - Menon (2007)')
```
